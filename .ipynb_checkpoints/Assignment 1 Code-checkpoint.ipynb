# 1.1

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from matplotlib.pyplot import figure

AV = [0,20,0,0,0,0]
IV = [30,45,60,40,40,40]
Ones = [15,20,25,15,20,25]
N=240

TC = np.zeros((240,6))

for i in range(len(AV)):
    TC[:,i] = np.zeros((1,240))
    start = AV[i]
    while start < (N-20) :
        for j in range(Ones[i]):
            TC[:,i][start+j] = np.ones(1)
        start += IV[i]
        
TC = preprocessing.scale(TC)
```

```python
fig, axs = plt.subplots(2,3, figsize = (15,10))

axs[0, 0].plot(TC[:,0])
axs[0, 0].set_title("TC 1")
axs[0, 1].plot(TC[:,1])
axs[0, 1].set_title("TC 2")
axs[0, 2].plot(TC[:,2])
axs[0, 2].set_title("TC 3")
axs[1, 0].plot(TC[:,3])
axs[1, 0].set_title("TC 4")
axs[1, 1].plot(TC[:,4])
axs[1, 1].set_title("TC 5")
axs[1, 2].plot(TC[:,5])
axs[1, 2].set_title("TC 6")

plt.savefig('question_1_1.jpg')
```

Standardize instead of Normalize if we Normalize then the mean will not be centered arund 0.

# 2.3

```python
nsrcs=6
N=240
V=441
x1=21
x2=21
x1x2=x1*x2
```

```python
p = np.arange(0,1.05,0.05)
step = 1/(np.linalg.norm(np.matmul(TC, TC.T), 'fro')*1.1)
avg_MSE = []

for val in p:
    MSE = []
    for i in range(10):
        temporal_noise=np.random.normal(0,sqrt(0.25),(240,6))
        spatial_noise=np.random.normal(0,sqrt(0.015),(6,441))
        X=np.matmul(TC+temporal_noise,SM+spatial_noise)
        X=preprocessing.scale(X)
        thr=val*N*step
        Ao=np.zeros((nsrcs,1))
        A=np.zeros((nsrcs,1))
        Alr=np.zeros((nsrcs,x1x2))
        
        for k in range(0,x1x2):
            
            sub_A =(X[:,k].reshape(X.shape[0],) - np.matmul(TC, Ao).reshape(X.shape[0],))
            A = Ao.reshape(Ao.shape[0],) + step*(np.matmul(TC.T, sub_A)).reshape(Ao.shape[0])
            A = (1/(1+thr))*np.sign(A)*np.maximum(np.repeat(0, nsrcs), abs(A)-thr)
            
            for i in range(0,10):
                Ao=A
                A = Ao + step*(np.matmul(TC.T, X[:,k] - np.matmul(TC, Ao)))
                A = (1/(1+thr)) * (np.sign(A) * np.maximum(np.repeat(0, nsrcs), abs(A) - thr))
            Alr[:,k] = A  
            
        Dlr=np.matmul(X,Alr.T)
        result = sum(sum((X - np.matmul(Dlr, Alr))**2)) / (N*V)
        MSE.append(result)
        
    avg_MSE.append(np.average(MSE))
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(p, avg_MSE)
```

```python
min(avg_MSE)
```

```python
p[avg_MSE.index(min(avg_MSE))]
```

Minimum MSE is found at 0.6 with an average MSE of 0.5456. It is okay to select this value for it is relatively in the middle betwen 0 and 1. MSE started to increase again or LR diverged again when rho is larger than 0.6 as can be seen from 0.65.

```python
for i in range(10):
        temporal_noise=np.random.normal(0,sqrt(0.25),(240,6))
        spatial_noise=np.random.normal(0,sqrt(0.015),(6,441))
        X=np.matmul(TC+temporal_noise,SM+spatial_noise)
        X=preprocessing.scale(X)
        thr=0.6*N*step
        Ao=np.zeros((nsrcs,1))
        A=np.zeros((nsrcs,1))
        Alr_rho=np.zeros((nsrcs,x1x2))
        
        for k in range(0,x1x2):
            
            sub_A =(X[:,k].reshape(X.shape[0],) - np.matmul(TC, Ao).reshape(X.shape[0],))
            A = Ao.reshape(Ao.shape[0],) + step*(np.matmul(TC.T, sub_A)).reshape(Ao.shape[0])
            A = (1/(1+thr))*np.sign(A)*np.maximum(np.repeat(0, nsrcs), abs(A)-thr)
            
            for i in range(0,10):
                Ao=A
                A = Ao + step*(np.matmul(TC.T, X[:,k] - np.matmul(TC, Ao)))
                A = (1/(1+thr)) * (np.sign(A) * np.maximum(np.repeat(0, nsrcs), abs(A) - thr))
            Alr_rho[:,k] = A  
            
        Dlr_rho=np.matmul(X,Alr_rho.T)
```

```python
C_trr=np.empty((6,1))
C_srr=np.empty((6,1))
C_tlr=np.empty((6,1))
C_slr=np.empty((6,1))

for i in range(6):
    corr = np.corrcoef(TC[:,i],Dlr_rho[:,i])
    C_tlr[i]=abs(corr[1,0])
print("C_tlr: ", sum(C_tlr)[0])

for i in range(6):
    corr = np.corrcoef(TC[:,i],D_rr[:,i])
    C_trr[i]=abs(corr[1,0])
print("C_trr: ", sum(C_trr)[0])

for i in range(6):
    corr = np.corrcoef(SM[i],Alr_rho[i])
    C_slr[i]=abs(corr[1,0])
print("C_slr: ", sum(C_slr)[0])

for i in range(6):
    corr = np.corrcoef(SM[i],A_rr[i])
    C_srr[i]=abs(corr[1,0])
print("C_srr: ", sum(C_srr)[0])
```
